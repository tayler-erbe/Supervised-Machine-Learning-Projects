```python
def remove_digits(strArr):
    return ''.join([i for i in strArr if not i.isdigit()])
```

```python
def remove_punctuation(str_Arr):
    # Removing punctuations in string
    # initializing punctuations string
    punc = '''!()-=+|[]{};:'"\,<>./?@#$%^&*_~'''
    # Using loop + punctuation string
    for ele in str_Arr:
        if ele in punc:
            str_Arr = str_Arr.replace(ele, "")
    
    return str_Arr
```

```python
#initialize 
lemmatizer = WordNetLemmatizer()
words = set(nltk.corpus.words.words())

def clean_text(strArr):  
    # Remove words not in dictionary
    remove_nonWords = " ".join(w for w in nltk.wordpunct_tokenize(strArr) \
                               if w.lower() in words or not w.isalpha())
    # Remove first word 'subject'
    remove_first_word = remove_nonWords.split(' ', 1)[1]
    # Remove Digits 
    fixed_digits = remove_digits(remove_first_word)
    # Expand contractions 
    fixed_contractions = expandContractions(fixed_digits, c_re=c_re)
    # Remove Punctuation
    fixed_puctuation = remove_punctuation(fixed_contractions)
    # Makes all lowercase, splits words by spaces
    split_lower = fixed_puctuation.lower().split()
    # Remove stop words
    remove_sw = [word for word in split_lower if not word in sw]
    # Lemmatize words
    cleaned_str = [lemmatizer.lemmatize(x) for x in remove_sw]
    
    # return result
    return cleaned_str
```
